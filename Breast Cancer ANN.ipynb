Step 1:

Loading the Data and libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder,StandardScaler
from sklearn.metrics import confusion_matrix as CM
from sklearn.metrics import accuracy_score,r2_score
from sklearn.neural_network import MLPClassifier,MLPRegressor
from sklearn.model_selection import train_test_split

from keras.models import Sequential
from keras.layers import Dense

import seaborn as sns
scaler = StandardScaler()
enc = LabelEncoder()
# Ignore warnings
#import warnings
#warnings.filterwarnings('ignore')
```

```python
df= pd.read_csv('data_refined.csv')
```

```python
df
```

```python
df = df.drop(df.iloc[:, 0:2], axis=1)
df
```

```python
X = df.drop('diagnosis_encoded', axis = 1)
X
```

```python
y= df['diagnosis_encoded']
y
```

Step 2:

Train Test Split

```python
train_ratio = 0.80
validation_ratio = 0.10
test_ratio = 0.10
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=1 - train_ratio)

# test is now 10% of the initial data set
# validation is now 10% of the initial data set
x_val, x_test, y_val, y_test = train_test_split(x_test, y_test, test_size=test_ratio/(test_ratio + validation_ratio)) 

print(x_train, x_val, x_test)
```

Step 3.1-3.3:

Sklearn ANN Classifier

```python
def SKClass():
    model = MLPClassifier(hidden_layer_sizes=(50,50,50), 
                      activation='relu', 
                      solver='adam', 
                      batch_size=20, 
                      max_iter=400, 
                      random_state=0,
                      verbose = False)    
    model.fit(x_train, y_train) 
    y1_pred = model.predict(x_test)
    y2_pred = model.predict(x_val)
    
    confusion_matrix1 = CM(y_test, y1_pred)
    confusion_matrix2 = CM(y_val, y2_pred)
    
    print('Accuracy score of test data: ' + str(accuracy_score(y_test,y1_pred)*100))
    print('Accuracy score of validation data: ' + str(accuracy_score(y_val,y2_pred)*100))
    
    
    
    return confusion_matrix1,confusion_matrix2
SKClass()
```

```python
def ANNClass():
    model = Sequential()

    model.add(Dense(units=50, activation='relu'))
    model.add(Dense(units=50, activation='relu'))
    model.add(Dense(units=50, activation='relu'))
    model.add(Dense(units=1, activation='relu'))


    model.compile(optimizer = 'adam',loss = "mean_squared_error",metrics = ['mean_squared_error'])

    model.fit(x_train, y_train, batch_size = 20, epochs = 20,verbose=False)
    y1_pred = model.predict(x_test)
    y2_pred = model.predict(x_val)

    y1_pred[np.where((y1_pred)>0.5)]=1
    y1_pred[np.where((y1_pred)<=0.5)]=0

    y2_pred[np.where((y2_pred)>0.5)]=1
    y2_pred[np.where((y2_pred)<=0.5)]=0

    confusion_matrix1 = CM(y_test, y1_pred)
    confusion_matrix2 = CM(y_val, y2_pred)

    print('Accuracy score of test data: ' + str(accuracy_score(y_test,y1_pred)*100))
    print('Accuracy score of validation data: ' + str(accuracy_score(y_val,y2_pred)*100))



    return confusion_matrix1,confusion_matrix2

ANNClass()
```

ANN's seem to get a better score of around 2-4% higher than the best Classifiers used before, however Keras Classifiers seem to be extremely in-efficient.

Step 6.1:

Sklearn ANN Regressor

```python
def InsSKReg():
    model = MLPRegressor(hidden_layer_sizes=(24,24,24), activation='relu',
                         solver='adam', batch_size=20, max_iter=70, random_state=0)
    model.fit(xins_train, yins_train) 
    y1_pred = model.predict(xins_test)
    y2_pred = model.predict(xins_val)
    
    score1 = r2_score(yins_test,y1_pred)
    score2 = r2_score(yins_val,y2_pred)
    
    return score1,score2

InsSKReg()
```

Keras ANN Regressor

```python
def InsKerReg():
    
    model = Sequential()
        
    model.add(Dense(units=30, activation='relu'))
    
    #Adding Hidden Layer
    model.add(Dense(units=30, activation='relu'))
    model.add(Dense(units=30, activation='relu'))
    
    #Adding Output Layer
    model.add(Dense(units=1, activation='relu'))
    
    model.compile(optimizer = 'adam', loss ='mean_squared_error', metrics = ['mean_squared_error'])
    model.fit(xins_train, yins_train, batch_size = 20, epochs = 20,verbose=False)
    y1_pred = model.predict(xins_test)
    y2_pred = model.predict(xins_val)
    
    score1 = r2_score(yins_test,y1_pred)
    score2 = r2_score(yins_val,y2_pred)
    
    return score1,score2
InsKerReg()
```
